```python
import pandas as pd    
```

```python
dataset = pd.read_csv('Startups-Merged-RatingsNullRemoved-DurationEncodedAsMonths-EmployeeStatusCategorical-Name-WorkSummary-HelpfulCount-ViewsCount-Date-Removed.csv')
```

```python
corpus=[]
for i in range(len(dataset)):
    Work_Detail = str(dataset['Work_Detail'][i]).split()
    Work_Detail=' '.join(Work_Detail)
    corpus.append(Work_Detail)
```

```python
print (corpus)
```

```python
Travel=[]
Timing_Nature=[]
Working_Days=[]
Rotational_Shift=[]
print(corpus[0].split(', ')[0])
print(len(corpus[0].split(', ')))
t=len(corpus)
print(t)
```

```python
for i in range(len(corpus)):
    #a=[]
    #corpus[i].split(', ')
    Travel.append('')
    Timing_Nature.append(' ')
    Working_Days.append(' ')
    Rotational_Shift.append('0')
    l=len(corpus[i].split(', '))
    #for j in range(0,l):
        #a=[]
       # a.append(corpus[i].split(', ')[j])
    for k in range(0,l):
        if(corpus[i].split(', ')[k]=='International Travel'):
            Travel[i]='2' 
        if(corpus[i].split(', ')[k]=='Travel Within Country'):
            Travel[i]='1'
        if(corpus[i].split(', ')[k]=='No Travel / Travel Within City'):
            Travel[i]='0'
        if(corpus[i].split(', ')[k]=='Strict Timings'):
            Timing_Nature[i]='1'
        if(corpus[i].split(', ')[k]=='Flexible Timings'):
            Timing_Nature[i]='0'
        if(corpus[i].split(', ')[k]=='Monday to Friday'):
            Working_Days[i]='0'
        
        if(corpus[i].split(', ')[k]=='Alternate Saturday off'):
            Working_Days[i]='1'
        
        if(corpus[i].split(', ')[k]=='Monday to Saturday'):
            Working_Days[i]='2'
        
        if(corpus[i].split(', ')[k]=='Rotational Shift'):
            Rotational_Shift[i]='1'
        
    
```

```python
print(Travel)
```

```python
df = pd.DataFrame(list(zip(corpus, Travel, Timing_Nature, Working_Days, Rotational_Shift)), 
               columns = ['Work_Detail', 'Travel', 'Timing_Nature', 'Working_Days', 'Rotational_Shift'])
```

```python
print(df)
```

```python
df.to_csv('Startups_Work_Details_Breakup.csv')
```
