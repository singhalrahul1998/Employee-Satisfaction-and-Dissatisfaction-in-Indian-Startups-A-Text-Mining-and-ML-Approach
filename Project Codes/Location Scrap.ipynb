```python
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
#Change path as needed
path_to_chromedriver = '/Users/LENOVO/Desktop/chromedriver/chromedriver.exe'
browser = webdriver.Chrome(executable_path = path_to_chromedriver)
import pandas as pd
```

```python
dataset = pd.read_csv('Startups-Merged-RatingsNullRemoved-DurationEncodedAsMonths-EmployeeStatusCategorical-Name-WorkSummary-HelpfulCount-ViewsCount-Date-Removed.csv')
```

```python
corpus=[]
for i in range(len(dataset)):
    Location = str(dataset['Location'][i]).split()
    Location=' '.join(Location)
    corpus.append(Location)
```

```python
print(corpus)
l=len(corpus)
```

```python
city = []
for i in range(len(corpus)):
    if(corpus[i]=='nan'):
        city.append('nan')
        continue
    url = 'https://www.google.com/maps/search/'+ corpus[i]+''
    browser.get(url)
    try:
        text=browser.find_element_by_xpath('//*[@id="pane"]/div/div[1]/div/div/div[2]/div[1]/div[1]/h1').get_attribute("innerHTML")
    except NoSuchElementException:
        text=browser.find_element_by_xpath('//*[@id="pane"]/div/div[1]/div/div/div[2]/div[1]/h2[2]/span').get_attribute("innerHTML")
    city.append(text)                       
```
