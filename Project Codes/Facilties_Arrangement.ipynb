```python
import pandas as pd
```

```python
dataset = pd.read_csv('Startups-Merged-RatingsNullRemoved-DurationEncodedAsMonths-EmployeeStatusCategorical-Name-WorkSummary-HelpfulCount-ViewsCount-Date-Removed.csv')
```

```python
corpus=[]
for i in range(len(dataset)):
    Facilities = str(dataset['Facilities'][i]).split()
    Facilities=' '.join(Facilities)
    corpus.append(Facilities)
```

```python
print (corpus)
```

```python
Free_transport=[]
Child_care=[]
Gymnasium=[]
Cafeteria=[]
Work_from_home=[]
Free_food=[]
Team_outings=[]
Education_assistance=[]
Soft_skill_training=[]
Health_insurance=[]
Job_training=[]
print(corpus[0].split(', ')[3])
print(len(corpus[0].split(', ')))
t=len(corpus)
print(t)
```

```python
for i in range(len(corpus)):
    #a=[]
    #corpus[i].split(', ')
    Free_transport.append('0')
    Child_care.append('0')
    Gymnasium.append('0')
    Cafeteria.append('0')
    Work_from_home.append('0')
    Free_food.append('0')
    Team_outings.append('0')
    Education_assistance.append('0')
    Soft_skill_training.append('0')
    Health_insurance.append('0')
    Job_training.append('0')
    l=len(corpus[i].split(', '))
    #for j in range(0,l):
        #a=[]
       # a.append(corpus[i].split(', ')[j])
    for k in range(0,l):
        if(corpus[i].split(', ')[k]=='Free transport'):
            Free_transport[i]='1'
        if(corpus[i].split(', ')[k]=='Child care'):
            Child_care[i]='1'
        
        if(corpus[i].split(', ')[k]=='Gymnasium'):
            Gymnasium[i]='1'
        
        if(corpus[i].split(', ')[k]=='Cafeteria'):
            Cafeteria[i]='1'
        
        if(corpus[i].split(', ')[k]=='Work from home'):
            Work_from_home[i]='1'
        
        if(corpus[i].split(', ')[k]=='Free food'):
            Free_food[i]='1'
        
        if(corpus[i].split(', ')[k]=='Team outings'):
            Team_outings[i]='1'
        
        if(corpus[i].split(', ')[k]=='Education assistance'):
            Education_assistance[i]='1'
        
        if(corpus[i].split(', ')[k]=='Soft skill training'):
            Soft_skill_training[i]='1'
       
        if(corpus[i].split(', ')[k]=='Health insurance'):
            Health_insurance[i]='1'
        if(corpus[i].split(', ')[k]=='Job training'):
            Job_training[i]='1'
    
```

```python
df = pd.DataFrame(list(zip(corpus, Free_transport, Child_care, Gymnasium, Cafeteria, Work_from_home, Free_food, Team_outings, Education_assistance, Soft_skill_training, Health_insurance, Job_training)), 
               columns = ['Facilities', 'Free_transport', 'Child_care', 'Gymnasium', 'Cafeteria', 'Work_from_home', 'Free_food', 'Team_outings', 'Education_assistance', 'Soft_skill_training', 'Health_insurance','Job_training'])
```

```python
print(df)
```

```python
df.to_csv('Startups_Work_Details_Breakup.csv')
```
